```julia
function f(x)
    a = 1;
    b = 1;
    for i=1:length(x)
        y = 0.3*sin(a)+0.4*b;
        z = 0.1*a+0.3*cos(b)+x[i];
        a = y;
        b = z;
    end
    return [a;b];
end
```

```julia
function ND(f)
    function g(x)
        # x = ones(2020);
        y = f(x); lx = length(x); ly = length(y);
        df = zeros(ly, lx);
        h = 1e-8;

        for i = 1 : lx
            x_tmp = ones(lx);
            x_tmp[i] = x_tmp[i] + h;
            y_tmp = f(x_tmp); 
            for j = 1 : ly
                df[j, i] = (y_tmp[j]-y[j])/(h);
            end
        end
        return df
    end
    return g;
end
```

```julia
@time df = ND(f);
```

```julia
N = 2020;
@time grad = df(ones(N,1));
println("The result is:")
println(grad * ones(N, 1))
```
